```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
corrr= df.corr()
```

```python
#Correlation with output variable
corrr_target = abs(corrr["churn"])
#Selecting highly correlated features
relevant_features = corrr_target[corrr_target>0.3]
relevant_features
```

```python
X_new = df[["tenure", "employ", "callcard", "loglong"]]
featureset = X_new

from sklearn.preprocessing import MinMaxScaler
x = featureset.values #returns a numpy array
min_max_scaler = MinMaxScaler()
feature_mtx = min_max_scaler.fit_transform(x)
X_new = feature_mtx
X_new
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2, max_iter=50000)
kmeans
kModel = kmeans.fit(X_new)
```

```python
pd.crosstab(Y, kModel.labels_)
```
