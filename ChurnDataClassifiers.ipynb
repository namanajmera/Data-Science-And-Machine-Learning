```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("datasets/ChurnData.csv")
data.corr()
```

```python
x = data[["tenure","employ" ]]
```

```python
y = data[["churn"]]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
x_train, x_test, y_train, y_test
```

# SVM

```python
from sklearn.svm import SVC
```

```python
clf = SVC(kernel='rbf').fit(x_train, y_train)
```

```python
y_hat = clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cnf = confusion_matrix(y_hat, y_test)
cnf
```

```python
cnf[0][0]+cnf[1][1]
```

```python
(31+0)/(31+9+0+0)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
obj = DecisionTreeClassifier(criterion="entropy")
obj = obj.fit(x_train,y_train)
obj_y_hat = obj.predict(x_test)
```

```python
obj_cnf = confusion_matrix(obj_y_hat, y_test)
```

```python
obj_cnf
```

```python
(22+7)/(22+2+7+9)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(max_iter = 500).fit(x_train,y_train)
yhat_prob = LR.predict(x_test)
yhat_prob
```

```python
lr_cnf = confusion_matrix(yhat_prob, y_test)
lr_cnf
```

```python
(29+3)/(29+3+6+2)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
k = 4
neigh = KNeighborsClassifier(n_neighbors = k).fit(x_train,y_train)
neigh
knn_yhat = neigh.predict(x_test)
```

```python
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh.predict(x_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, knn_yhat))
```
