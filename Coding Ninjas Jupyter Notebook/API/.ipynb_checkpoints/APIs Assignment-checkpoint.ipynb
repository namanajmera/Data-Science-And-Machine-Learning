```python
#Print all sub-breeds of hound by making a get request to given URL.
#URL : https://dog.ceo/api/breeds/list/all
#Print name of sub-breeds in different lines.

import requests
import json
p=requests.get("https://dog.ceo/api/breeds/list/all")

python_data=p.json()
for i in python_data['message']['hound']:
    print(i)
```

```python
#Print name of all online courses offered by Coding Ninjas. 
#You can fetch the data using their open API :
#https://api.codingninjas.com/api/v3/courses .

import requests
import json
p=requests.get("https://api.codingninjas.com/api/v3/courses")

python_data=p.json()
all_courses=python_data['data']['courses']
for i in all_courses:
    print(i['name'])
```

```python
#Count the total number of cities of country AU from this 
#URL : https://api.openaq.org/v1/cities/

import requests
import json

json_data=requests.get("https://api.openaq.org/v1/cities",params={"country":"AU"})

python_data=json_data.json()
cities=python_data['results']
print(len(cities))
```

```python
#Print first 5 cities from country “BE”. Fetch the result from given URL.
#Order the results by city.
#URL - https://api.openaq.org/v1/cities
#Documentation - https://docs.openaq.org/

import requests
import json

json_data=requests.get("https://api.openaq.org/v1/cities",params={"country":"BE","limit":5,"order_by":"city"})

python_data=json_data.json()
cities=python_data['results']
for i in cities:
    print(i['city'])
```

Fetch the list of contributors in the repository 'CodingNinjasCodes/JSNotes' 

in Github from Github API using OAuth(access token).

Print the username of contributors and its number of commit done.

```python
import requests
import json
h={'accept':'application/vnd.github.v3+json',
   'Authorization':'token e67585b335eed4ace4aa26adb44de5871c0f1a2e'}
fetch=requests.get('https://api.github.com/repos/CodingNinjasCodes/JSNotes/stats/contributors',headers=h,params={'type':'all'})
data=fetch.json()
for i in data:
    print(i['author']['login'],i['total'])
```
