# Bubble Chart

```python
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 300, 370]
```

```python
plt.scatter(year,salary,s=population,c=population) # Bubble size to signify population
plt.show()
```

```python
#assign the text of population
plt.scatter(year,salary,s=population,c=population)

for i in range(len(salary)):
    plt.text(year[i]+0.1,salary[i]+0.1,population[i])
    
plt.show()
```

```python
#alpha define the transperance of plotting

plt.scatter(year,salary,s=population,c=population,alpha=0.5)

for i in range(len(salary)):
    plt.text(year[i]+0.1,salary[i]+0.1,population[i])
    
plt.show()
```

```python
#edgde color of each bubble
plt.scatter(year,salary,s=population,c=population,alpha=5,edgecolor='red')
plt.title("Bubble Chart")
for i in range(len(salary)):
    plt.text(year[i]+0.1,salary[i]+0.1,population[i])
    
plt.show()
```

```python
#Label in Pie Chart
size=[3,4,6,2]
labels=["A","B","C","D"]
plt.pie(size,labels=labels)
plt.axis("equal")
plt.show()
```

```python
#change the color in pie chart
size=[3,4,6,2]
labels=["A","B","C","D"]
colors=['blue','red','pink','purple']
plt.title("Split among Classes")
plt.pie(size,labels=labels,colors=colors)
plt.axis("equal")
plt.show()
```

```python
#label inside the pie chart with percentage
size=[3,4,6,2]
labels=["A","B","C","D"]
colors=['blue','red','pink','purple']
plt.title("Split among Classes")
plt.pie(size,labels=labels,colors=colors,autopct='%.2f%%')  #we can also pass the function in autopct
plt.axis("equal")
plt.show()
```

```python
#explode the pie chart
size=[3,4,6,2]
labels=["A","B","C","D"]
colors=['blue','red','pink','purple']
plt.title("Split among Classes")
explode=[0.1,0.1,0.1,0]
plt.pie(size,labels=labels,colors=colors,explode=explode,autopct='%.2f%%')
plt.axis("equal")
plt.show()
```

```python
#plot in clock wise
size=[3,4,6,2]
labels=["A","B","C","D"]
colors=['blue','red','pink','purple']
plt.title("Split among Classes")
explode=[0.1,0.1,0.1,0]
plt.pie(size,labels=labels,colors=colors,explode=explode,autopct='%.2f%%',counterclock=False) #by default it is true
plt.axis("equal")
plt.show()
```

```python
#set the start angle
size=[3,4,6,2]
labels=["A","B","C","D"]
plt.title("Split among Classes")
colors=['blue','red','pink','purple']
explode=[0.1,0.1,0.1,0]
plt.pie(size,labels=labels,colors=colors,explode=explode,autopct='%.2f%%',counterclock=False,startangle=90) #by default it is 0 degree
plt.axis("equal")
plt.show()
```

```python
#Laptop Sales

import matplotlib.pyplot as plt

company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]

plt.pie(sold,autopct='%.1f')
plt.show()

sumSold=sum(sold)
l=[]
for i in sold:
    l.append((i/sumSold)*100)

for i in range(len(company)):
    print(company[i],'%.1f'%l[i])
```

```python
# show the percentage and value both

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct

plt.pie(sold, labels=company, autopct=make_autopct(sold))
plt.show()
```

```python
# By default 10 bins are made

# change number of bins


# width of each bin is (total bins) / (number of bins)

# bins follow [a, b) convetion (a included, b not included)

# in last both are included

xt=np.arange(23)
bins=len(b)
plt.hist(b,bins=bins,edgecolor='black')
plt.xticks(xt)
plt.show()
```

```python
#also we can specify the bins

plt.hist(b, bins = [1, 3, 6, 10, 15, 20,23], edgecolor = "black",orientation='horizontal')
plt.yticks(xt)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("amazon_jobs_dataset.csv")
df_posting_date=df['Posting_date'].str.split(', ').str[1]

year={}
for i in df_posting_date:
    if i in year:
        year[i]=year[i]+1
    else:
        year[i]=1

x=[]
y=[]

for i in year:
    x.append(int(i))
    y.append(year[i])
    
plt.plot(x,y)
plt.show()

l=len(x)-1

while l>=0:
    print(x[l],y[l])
    l-=1
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("amazon_jobs_dataset.csv")
df_posting_date=df['Posting_date'].str.split(', ').str[0]
df_month=df_posting_date.str.split(' ').str[0]

month={}
for i in df_month:
    if i in month:
        month[i]=month[i]+1
    else:
        month[i]=1

x=[]
y=[]
for i in month:
    x.append(i)
    y.append(month[i])

plt.bar(x,y)
plt.xticks(rotation=70)
plt.show()

for i in range(len(x)):
    print(x[i],y[i])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("amazon_jobs_dataset.csv")
df_city=df[df['location'].str.split(',').str[0]=='IN']
df_city=df_city['location'].str.split(',').str[2]

city={}
for i in df_city:
    if i in city:
        city[i]=city[i]+1
    else:
        city[i]=1
x=[]
y=[]
for i in city:
    x.append(i)
    y.append(city[i])
    
explodes=[0.01,0.01,0.01,0.2,0.5]
plt.pie(y,autopct='%.2f%%',labels=x,explode=explodes)
plt.show()

y_percen=[]
sumy=sum(y)
for i in range(len(y)):
    y_percen.append((y[i]/sumy)*100)
    print(x[i],format(y_percen[i], '.2f')) 
```

```python
import csv

with open('amazon_jobs_dataset.csv', encoding='utf8') as file_obj:
    file_data=csv.DictReader(file_obj, skipinitialspace=True)
    dic=dict()
    city=[]
    for row in file_data:
        if 'IN' in row['location'] and not 'IE' in row['location']:
            city.append(row['location'].split()[2])
    for i in city:
        if i in dic:
            dic[i]+=1
        else:
            dic[i]=1
            
    xaxis=[]
    yaxis=[]
    for i in dic:
        xaxis.append(i)
        yaxis.append(dic[i])
    np_xaxis=np.array(xaxis)
    np_yaxis=np.array(yaxis, dtype='int')
    
    np_xaxis=np_xaxis[np.argsort(np_yaxis)]
    np_yaxis=np.sort(np_yaxis)
    
    np_xaxis=np_xaxis[::-1]
    np_yaxis=np_yaxis[::-1]
    
    plt.pie(np_yaxis, labels=np_xaxis, autopct='%.2f%%', explode=[0.1, 0.1, 0.1, 0.1, 0.8])
    plt.show()
    
    for i in range(len(np_xaxis)):
        print(np_xaxis[i], format((np_yaxis[i]*100)/sum(dic.values()), '.2f'))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

with open('amazon_jobs_dataset.csv', encoding='utf8') as file_obj:
    file_data=csv.DictReader(file_obj, skipinitialspace=True)
    year=[]
    for row in file_data:
        if 'java' in row['BASIC QUALIFICATIONS'] or 'Java' in row['BASIC QUALIFICATIONS']:
            year.append(row['Posting_date'].split()[2])
    np_year=np.array(year, dtype='int')

year={}
for i in np_year:
    if i in year:
        year[i]=year[i]+1
    else:
        year[i]=1
x=list(year.keys())
y=list(year.values())

plt.scatter(x,y)
plt.show()

l=len(x)-1

while l>=0:
    print(x[l],y[l])
    l-=1
```
