```python
#Find the data of flower “Iris-virginica” type where petal-length > 1.5?

import pandas as pd

def print_df(df):
    df_ = df.copy()
    df_.columns = ['', '', '', '','']
    df_.index = ['' for _ in range(len(df))]
    #print(len(df))
    print(df_)

iris=pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")

for i in iris.index:
    if iris.loc[i,'variety']=='Virginica':
        iris.loc[i,'variety']='Iris-virginica'
#print(iris)

iris=iris[iris['variety']=='Iris-virginica']
iris=iris[iris['petal.length']>1.5]
print_df(iris)
```

```python
#Find and print the minimum, maximum and average value of the feature for each kind of flower ?
#Note: Get the dataset from here
#Print the value with two decimal places.
#Note: Order for flower is Iris-setosa, Iris-versicolor and Iris-virginica.

import pandas as pd 
column_name=['SeapalLength','SepalWidth','PetalLength','PetalWidth','Species']
iris=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",names=column_name)
df=iris.copy()
setosa=df[df.Species=='Iris-setosa'].describe()
virginica=df[df.Species=='Iris-virginica'].describe()
versicolor=df[df.Species=='Iris-versicolor'].describe()

setosamin=setosa.loc['min']
setosamax=setosa.loc['max']
avgSLS=setosa.loc['mean']

virginicamin=virginica.loc['min']
virginicamax=virginica.loc['max']
avgSLVI=virginica.loc['mean']

versicolormin=versicolor.loc['min']
versicolormax=versicolor.loc['max']
avgSLVE=versicolor.loc['mean']

print('%.2f'%setosamin['SeapalLength'],'%.2f'%setosamin['SepalWidth'],'%.2f'%setosamin['PetalLength'],'%.2f'%setosamin['PetalWidth'],"Iris-setosa")
print('%.2f'%setosamax['SeapalLength'],'%.2f'%setosamax['SepalWidth'],'%.2f'%setosamax['PetalLength'],'%.2f'%setosamax['PetalWidth'],"Iris-setosa")
print('%.2f'%avgSLS['SeapalLength'],'%.2f'%avgSLS['SepalWidth'],'%.2f'%avgSLS['PetalLength'],'%.2f'%avgSLS['PetalWidth'],"Iris-setosa")

print('%.2f'%versicolormin['SeapalLength'],'%.2f'%versicolormin['SepalWidth'],'%.2f'%versicolormin['PetalLength'],'%.2f'%versicolormin['PetalWidth'],"Iris-versicolor")
print('%.2f'%versicolormax['SeapalLength'],'%.2f'%versicolormax['SepalWidth'],'%.2f'%versicolormax['PetalLength'],'%.2f'%versicolormax['PetalWidth'],"Iris-versicolor")
print('%.2f'%avgSLVE['SeapalLength'],'%.2f'%avgSLVE['SepalWidth'],'%.2f'%avgSLVE['PetalLength'],'%.2f'%avgSLVE['PetalWidth'],"Iris-versicolor")

print('%.2f'%virginicamin['SeapalLength'],'%.2f'%virginicamin['SepalWidth'],'%.2f'%virginicamin['PetalLength'],'%.2f'%virginicamin['PetalWidth'],"Iris-virginica")
print('%.2f'%virginicamax['SeapalLength'],'%.2f'%virginicamax['SepalWidth'],'%.2f'%virginicamax['PetalLength'],'%.2f'%virginicamax['PetalWidth'],"Iris-virginica")
print('%.2f'%avgSLVI['SeapalLength'],'%.2f'%avgSLVI['SepalWidth'],'%.2f'%avgSLVI['PetalLength'],'%.2f'%avgSLVI['PetalWidth'],"Iris-virginica")

```

```python
# Python code to demonstrate precision 
# and round() 
  
# initializing value 
a = 3.4
  
# using "%" to print value till 2 decimal places  
print ("The value of number till 2 decimal place(using %) is : ",end="") 
print ('%.2f'%a) 
  
# using format() to print value till 2 decimal places  
print ("The value of number till 2 decimal place(using format()) is : ",end="") 
print ("{0:.2f}".format(a)) 
  
# using round() to print value till 2 decimal places  
print ("The value of number till 2 decimal place(using round()) is : ",end="") 
print (round(a,2)) 
```

```python
#Find out the Country with Highest Number of Terror Attack 
#and in which year the most number of terrorist attack happened in that country ?

import pandas as pd
import numpy as np

df=pd.read_csv("terrorismData.csv")
df=df[df.Country==df.Country.describe().top]
count=df.shape[0]
country=df.Country.describe().top

y={}

for i in df.Year:
    if i in y:
        y[i]+=1
    else:
        y[i]=1
cnt=0
year=0
for i in y:
    if cnt<y[i]:
        cnt=y[i]
        year=i
print(country,count,year)
```
